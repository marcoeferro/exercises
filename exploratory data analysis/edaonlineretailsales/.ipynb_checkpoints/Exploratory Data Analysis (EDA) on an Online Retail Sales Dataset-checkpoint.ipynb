# RFM

## RFM (Recency, Frequency, Monetary) Analysis

RFM analysis is a marketing technique used to segment customers based on their purchasing behavior. The goal is to identify the most valuable customers by evaluating three key metrics:

- **Recency (R):** How recently a customer made a purchase.
- **Frequency (F):** How often a customer makes a purchase.
- **Monetary (M):** How much money a customer spends on purchas#es.

## Steps to Conduct RFM Ana#lysis

### 1. Data Collection

- Gather transactional data from your sales records, including customer IDs, purchase dates, and purchas#e amounts.

### 2. Data Preparation

- Clean the data to ensure accuracy. Remove duplicates and correct any errors.
- Ensure that each transaction record includes the customer ID, transaction date, and trans#action amount.

### 3. Calculate RFM Metrics

- **Recency (R):** Calculate the number of days since the last purchase for each customer.
  - `Recency = Today's Date - Date of Last Purchase`
- **Frequency (F):** Count the total number of purchases made by each customer within a specified period.
  - `Frequency = Total Number of Purchases`
- **Monetary (M):** Sum the total amount of money spent by each customer within a specified period.
  - `Monetary = S#um of Purchase Amounts`

### 4. Score Assignment

- Divide customers into groups for each RFM metric. Commonly, this is done by assigning scores from 1 to 5, where 5 represents the highest value (most recent, most frequent, highest spending).
  - For example, the top 20% of customers in terms of recency might receive a score of 5, the next 20% a score of 4, and so on.
- This results in each customer having three scores (one for Recency, one for Freq##ency, and one for Monetary).

### 5. RFM Segmentation

- Combine the three scores to create a composite RFM score. For example, a customer with scores of R=5, F=4, and M=3 would have an RFM score of 543.
- Group customers based on their RFM scores. Customers with similar RFM scores exhibit similar behaviors and can be targeted wi#th tailored marketing strategies.

### 6. Analyze and Interpret

- Identify high-value customer segments, such as those with high scores in all three metrics.
- Develop specific marketing strategies for different segments. For instance, high recency and frequency but low monetary value customers might be encouraged to make higher-value purchases.

## RFM Scores (Recency, Frequency, Monetary)

This section details the RFM scoring system used to evaluate customer behavior. Scores range from 1 (worst) to 5 (best).

**Recency Score:**

* **High Score (5):** Indicates a recent customer interaction (good).
* **Low Score (1):** Indicates a long period since the last interaction (bad).

**Frequency Score:**

* **High Score (5):** Represents customers who frequently interact with the company (good).
* **Low Score (1):** Represents customers with infrequent interactions (bad).

**Monetary Score:**

* **High Score (5):** Represents customers with high spending (good).
* **Low Score (1):** Represents customers with low spending (bad).

**In summary:**

* We want customers with:
    * Recent interactions (high recency score).
    * Frequent interactions (high frequency score).
    * High spending (high monetary score).
ny.ny.any.ompany.

```python
# Create the recency_score, frequency_score, and monetary_score values.
rfm_df["Recency_Score"] = pd.qcut(rfm_df["Recency"], 5, labels = [5, 4, 3, 2, 1]) 
rfm_df["Frequency_Score"] = pd.qcut(rfm_df["Frequency"].rank(method = "first"), 5, labels = [1, 2, 3, 4, 5])
rfm_df["Monetary_Score"] = pd.qcut(rfm_df["MonetaryValue"], 5, labels = [1, 2, 3, 4, 5])
```

```python
# Last version of df_rfm dataframe.
rfm_df.head()
```

```python
# I create the RF_score using the recency_score and frequency_score.
rfm_df["RF_Score"] = rfm_df["Recency_Score"].astype(str) + rfm_df["Frequency_Score"].astype(str)
```

```python
# Last version of df_rfm dataframe.
rfm_df.head()
```

```python
# Seg_map is customer group.
seg_map = {
    r'[1-2][1-2]' : 'hibernatig',
    r'[1-2][3-4]' : 'at_Risk',
    r'[1-2]5' : 'cant-loose',
    r'3[1-2]' : 'about_to_sleep',
    r'33' : 'need_attention',
    r'[3-4][4-5]' : 'loyal_customers',
    r'41' : 'promising',
    r'51' : 'new_customers',
    r'[4-5][2-3]' : 'potential_loyalists',
    r'5[4-5]' : 'champions'}
```

```python
# I create segment using RF_Score and segmap.
rfm_df["segment"] = rfm_df["RF_Score"].replace(seg_map, regex = True)
```

```python
# Last version of df_rfm dataframe.
rfm_df.head()
```

# CLTV
**CLTV**, or **Customer Lifetime Value**, is a metric that estimates the total revenue a company can expect to generate from a customer over the entire business relationship with that customer. It is a measure of how much value a customer brings to the company over time.

## Calculating and subtracting outlier values from the data

```python
def outlier_thresholds(dataframe, variable):
    quartile1 = dataframe[variable].quantile(0.01)
    quartile3 = dataframe[variable].quantile(0.99)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return up_limit, low_limit

def replace_with_threshold(dataframe, variable):
  up_limit, low_limit = outlier_thresholds(dataframe, variable)

  # Ensure column exists before filtering
  if variable not in dataframe.columns:
    raise ValueError(f"Variable '{variable}' not found in the DataFrame.")

  # Efficient filtering using boolean indexing
  return dataframe.query(f"{variable} > @low_limit & {variable} < @up_limit")
```

## Prepare CLTV DataFrame

```python
# It is the last date in the dataset. I will need this to calculate the recency values. I'm going to use this to specify the today_date value.
data["InvoiceDate"].max() #Timestamp('2011-12-09 12:50:00')
```

**Recency:** Time since the last purchase. Weekly. (User Specific)

**T:** Customer's age. Weekly. (how long before the analysis date the first purchase was made)

**Frequency:** Total number of repeat purchases (frequency > 1)

**Monetary:** Average earnings per purchase

```python
today_date = dt.datetime(2011,12,10) #i will assume that only one day has passed 

# Create the df_cltv dataframe.
df_cltv = data.groupby("Customer ID").agg({"InvoiceDate" : [lambda InvoiceDate: (InvoiceDate.max() - InvoiceDate.min()).days, # recency
                                                  lambda InvoiceDate: (today_date - InvoiceDate.min()).days], # T (total observation time per client)
                                         "Invoice" : lambda Invoice: Invoice.nunique(), # frequnecy
                                         "Total_Prize": lambda Total_Prize: Total_Prize.sum()}) # monatery
```

```python
# First version of the df_cltv dataframe.
df_cltv.head()
```

```python
# I remove the column names of df_cltv because I am going to change them.
df_cltv.columns = df_cltv.columns.droplevel(0)
```

```python
df_cltv.head()
```

```python
# I specfy the column name of df_cltv
df_cltv.columns = ["recency", "T", "frequency", "monetary"]
```

```python
df_cltv.head()
```

```python
# Average earnings per purchase
df_cltv["monetary"] = df_cltv["monetary"] / df_cltv["frequency"]
```

```python
df_cltv.head()
```

```python
# any client with less than 1 purches will be discarded
df_cltv = df_cltv[df_cltv["frequency"] > 1]
```

```python
df_cltv.head()
```

## BG-NBD

In this section, I will apply the BetaGeoFitter model to the df_cltv so i can predict the purchased rateexpected in a period of time for each customer

```python
from lifetimes import BetaGeoFitter
from lifetimes import GammaGammaFitter
from lifetimes.plotting import plot_period_transactions
from sklearn.preprocessing import MinMaxScaler
```

```python
# Model
bgf = BetaGeoFitter(penalizer_coef = 0.001)

# fit with frequency, recency, and T values.
bgf.fit(df_cltv["frequency"],
       df_cltv["recency"],
       df_cltv["T"])
```

```python
# Calculating one monthly values
bgf.predict(30,
           df_cltv["frequency"],
           df_cltv["recency"],
           df_cltv["T"]).sort_values(ascending = False).head(10)
```

```python
# Calculating three monthly values
bgf.predict(90, 
           df_cltv["frequency"],
           df_cltv["recency"],
           df_cltv["T"]).sort_values(ascending = False).head(10)
```

```python
# Adding the three monthly values to the df_cltv
df_cltv["expected_purc_3_month"] = bgf.predict(90, 
                                               df_cltv["frequency"],
                                               df_cltv["recency"],
                                               df_cltv["T"])
```

```python
df_cltv.head()
```

```python
# Calculating six monthly values
bgf.predict(180, 
           df_cltv["frequency"],
           df_cltv["recency"],
           df_cltv["T"]).sort_values(ascending = False).head(10)
```

```python
# Adding the six monthly values to the df_cltv
df_cltv["expected_purc_6_month"] = bgf.predict(180, 
                                               df_cltv["frequency"],
                                               df_cltv["recency"],
                                               df_cltv["T"])
```

```python
df_cltv.head()
```

## Gamma Gamma

```python
# Model
ggf = GammaGammaFitter(penalizer_coef = 0.01)

# fitting with frequency adn monetary values
ggf.fit(df_cltv["frequency"], df_cltv["monetary"])
```

At the code block below, I calculate how much profit a customer will bring in with the gammagammafitter model.

```python
ggf.conditional_expected_average_profit(df_cltv["frequency"], df_cltv["monetary"]).sort_values(ascending = False).head(10)
```

```python
# Adding the prediction values to df_rfm.
df_cltv["expected_average_profit"] = ggf.conditional_expected_average_profit(df_cltv["frequency"], df_cltv["monetary"])
```

```python
df_cltv.head()
```

At the code block below, I calculate the cltv values of the customer.

```python
cltv = ggf.customer_lifetime_value(bgf,
                                  df_cltv["frequency"],
                                  df_cltv["recency"],
                                  df_cltv["T"],
                                  df_cltv["monetary"],
                                  time = 6, # six mounth
                                  freq = "W",
                                  discount_rate = 0.01)
```

```python
cltv.head(10)
```

```python
# Adding the cltv values to the df_cltv.
df_cltv["cltv"] = cltv
```
